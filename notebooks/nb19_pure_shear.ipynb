# Notebook 19: Progressive strain

So far, we have focused on the undeformed (initial) and deformed (final) states. However, finite strain is the cumulative result of a series of strain increments:

$$
\begin{aligned}
&{ }^1 x_i={ }^1 F_{i j}{ }^1 X_j \\
&{ }^2 X_i={ }^1 x_i \\
&{ }^2 x_i={ }^2 F_{i j}{ }^2 X_j \\
&{ }^3 X_i={ }^2 x_i \\
&{ }^3 x_i={ }^3 F_{i j}{ }^3 X_j \\
& \cdots \\
&{ }^n x_i={ }^n F_{i j}{ }^n X_j
\end{aligned}
$$

where $F_{ij}$ is the Green deformation gradient, and 1 to $n$ are the strain increments. This can also be written as:

$$
x_i={ }^n F_{i j} \ldots{ }^3 F_{i j}{ }^2 F_{i j}{ }^1 F_{i j}{ }^1 X_j
$$

Notice that since ${ }^2 \mathbf{F}^1 \mathbf{F} \neq{ }^1 \mathbf{F}^2 \mathbf{F}$, for finite strain we must know the order of the deformation. For example, if we want to determine the finite strain of a group of faults in a region, we must know the order in which these faults occurred. This is often difficult to determine.

In this and the next two notebooks, we will look at some simple deformations, which are characterized by the same incremental deformation gradient through time. For simplicity, we will assume that there is no strain along the $\mathbf{X}_2$ axis. Thus, we are dealing with plane strain.

## Pure shear:

The first type of deformation we will cover is pure shear. For this deformation, the principal stretches are along the coordinate axes. Letâ€™s assume the maximum principal stretch, $S_1$, is along $\mathbf{X}_1$, and the minimum principal strecth, $S_3$, is along $\mathbf{X}_3$. $S_2 = 1$ and is parallel to $\mathbf{X}_2$. The Green deformation gradient for this case is ([Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)):

$$
{ }^{P S} F_{i j}=\left[\begin{array}{ccc}
S_1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & S_3
\end{array}\right]
$$

## Python function

The function [pure_shear](../functions/pure_shear.py) deforms a collection of points using pure shear and a maximum stretch, $S_1$. The function displays on the left the undeformed (blue) and deformed (red) geometries (and by default the displacement paths), and on the right the value of $S_1$ versus the angle, $\Theta$, $S_1$ makes with the $\mathbf{X}_1$ axis.

## Application:

Let's use this function to deform a square using pure shear:

```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt

# this makes visible our functions folder
sys.path.append(os.path.abspath(os.path.join("..", "functions")))

# import pure_shear function
from pure_shear import pure_shear

# initial points coordinates
pts = np.zeros((16,2)) 
pts[:,0]=[-1,-1,-1,-1,-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5] 
pts[:,1]=[-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5,-1,-1,-1,-1]

# maximum stretch
S1 = 2.5

# number of increments
ninc = 10

# deform and plot
fig,ax = plt.subplots(1,2,figsize=(15,5)) 
paths, psf = pure_shear(pts,S1,ninc,ax) 
plt.show()
```

$S_1$ and $S_3$ do not rotate throughout the deformation ($\Theta$ is 0 throughout the deformation). Pure shear is a *non-rotational* deformation.
