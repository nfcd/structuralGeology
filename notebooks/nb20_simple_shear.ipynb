# Notebook 20: Simple shear

For plane strain and simple shear along the $\mathbf{X}_1$ axis, the Green deformation gradient is:

$$
{ }^{S S} F_{i j}=\left[\begin{array}{lll}
1 & 0 & \gamma \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right]
$$

where $\gamma$ is the shear strain.

## Python function

The function [simple_shear](../functions/simple_shear.py) deforms a collection of points using simple shear and a value of shear strain, gamma. The function displays on the left the undeformed (blue) and deformed (red) geometries (and by default the displacement paths), and on the right the value of $S_1$ versus the angle $\Theta$, $S_1$ makes with the $\mathbf{X}_1$ axis.

## Application:

Let's use this function to deform a square using simple shear:

```python
import sys, os
import numpy as np
import matplotlib.pyplot as plt

# this makes visible our functions folder
sys.path.append(os.path.abspath(os.path.join("..", "functions")))

# import simple_shear function
from simple_shear import simple_shear

# initial points coordinates
pts = np.zeros((16,2)) 
pts[:,0]=[-1,-1,-1,-1,-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5] 
pts[:,1]=[-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5,-1,-1,-1,-1]

# shear strain
gamma = 2.5

# number of increments
ninc = 10

# deform and plot
fig,ax = plt.subplots(1,2,figsize=(15,5)) 
paths,psf = simple_shear(pts,gamma,ninc,ax) 
plt.show()
```

In this case, the principal strain axes, $S_1$ and $S_3$, rotate throughout the deformation. $\Theta$ is initially 45° (as predicted by the theory of infinitesimal strain), and it progressively decreases to a value of 20° by the end of the deformation. Simple shear is a *rotational* deformation.
