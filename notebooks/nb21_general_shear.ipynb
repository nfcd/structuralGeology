# Notebook 21: General shear

General shear is a combination of pure shear and simple shear. If the shear direction and $S_1$ are parallel to $\mathbf{X}_1$, $S_2 = 1$ (plane strain), and area is constant, the Green deformation gradient is (Allmendinger et al., 2012):

$$
{ }^{G S} F_{i j}=\left[\begin{array}{ccc}
S_1 & 0 & \frac{\gamma\left(S_1-S_3\right)}{2 \ln S_1} \\
0 & 1 & 0 \\
0 & 0 & S_3
\end{array}\right]
$$

If the shear direction is parallel to $\mathbf{X}_1$, but $S_1$ is parallel to $\mathbf{X}_3$, the Green deformation gradient is (Allmendinger et al., 2012):

$$
{ }^{G S} F_{i j}=\left[\begin{array}{llc}
S_3 & 0 & \frac{\gamma\left(S_3-S_1\right)}{2 \ln S_3} \\
0 & 1 & 0 \\
0 & 0 & S_1
\end{array}\right]
$$

A simple dimensionless measure of the ratio of simple to pure shear is the cosine of the acute angle between the eigenvectors of the Green deformation gradient $\mathbf{F}$. This measure is known as the *kinematic vorticity number*, $W_k$ (Truesdell, 1953). General shear is characterized by a $W_k$ between 0 (pure shear) and 1 (simple shear).

## Python function

The function [general_shear](../functions/general_shear.py) deforms a collection of points using general shear and values of $S_1$ (S1) and $\gamma$ (gamma), for a shear direction parallel to $\mathbf{X}_1$, and $S_1$ parallel (kk = 0) or perpendicular to the shear direction (kk = 1). [general_shear](../functions/general_shear.py) displays the points' displacement paths for a number of increments (ninc), and $S_1$ versus $\Theta$.

## Application:

Let's use this function to deform a square using general shear, for a shear direction and $S_1$ parallel to $\mathbf{X}_1$:

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

# this makes visible our functions folder
sys.path.append(os.path.abspath(os.path.join('..', 'functions')))

# import general_shear function
from general_shear import general_shear

# initial points coordinates
pts = np.zeros((16,2)) 
pts[:,0]=[-1,-1,-1,-1,-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5] 
pts[:,1]=[-1,-0.5,0,0.5,1,1,1,1,1,0.5,0,-0.5,-1,-1,-1,-1]

# S1 and gamma
S1 = 2.0
gamma = 0.5

# number of increments
ninc = 10

# S1 parallel to shear direction
kk = 0

# deform and plot
fig,ax = plt.subplots(1,2,figsize=(15,5))
paths1, wk1, psf1 = general_shear(pts,S1,gamma,kk,ninc,ax) 
print(f"Wk = {wk1:.4f}")
plt.show()
```

Now let's make $S_1$ perpendicular to the shear direction:

```python
# S1 perpendicular to shear direction
kk = 1

# deform and plot
fig,ax = plt.subplots(1,2,figsize=(15,5))
paths1, wk1, psf1 = general_shear(pts,S1,gamma,kk,ninc,ax) 
print(f"Wk = {wk1:.4f}")
plt.show()
```

In both cases, the contribution of simple shear is about one third that of pure shear ($W_k$ = 0.34). Try running the notebook with different values of $S_1$ and $\gamma$, and check $W_k$ and $\Theta$.
