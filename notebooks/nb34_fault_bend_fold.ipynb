# Notebook 34: Fault bend folds

In sedimentary basins, folds are often associated with faults. Notebooks 34 to 37 are about fault related folds and kinematic (velocity) models to simulate them.

We start with fault bend folds, where the fold is formed by movement of the rocks above a non-planar fault ([Suppe, 1983](https://ajsonline.org/article/60277-geometry-and-kinematics-of-fault-bend-folding)). The figure below (from [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)) shows the kinematic model for a simple fault step, fault bend fold. Note that the model consists of three velocity domains, $v_1$ and $v_2$ are the velocities along the $\mathbf{x_1}$ and $\mathbf{x_2}$ axes, $s$ is the slip, and $R$ is a slip ratio (more details in [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)):

<img src="../figures/fault_bend_fold.png" alt="fault_bend_fold" width="500" style="display: block; margin: 0 auto"/><br><br>

Fault bend folds grow as a result of kink band migration during two stages ([Suppe, 1983](https://ajsonline.org/article/60277-geometry-and-kinematics-of-fault-bend-folding)). The figure below (from [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)) shows these two stages. In the first stage (a), the displacement is less than the length of the fault ramp. The anticline increases in height and the crest decreases in width. In the second stage (b), the displacement is greater than the length of the ramp. The crest of the anticline increases in width but stops growing in height. This becomes clear by looking at the evolution of a fault bend fold.

<img src="../figures/fbf_two_stages.png" alt="fbf_two_stages" width="500" style="display: block; margin: 0 auto"/><br><br>

The function [fault_bend_fold](../functions/fault_bend_fold.py) plots the evolution of a simple fault step, fault bend fold. The cell below set the input parameters and calls the function. Note that the two stages of the fold are labeled, pay attention when stage 1 changes to stage 2.

```python
import numpy as np

# import our function
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "functions")))
from fault_bend_fold import fault_bend_fold

# define the input parameters
# beds datums
yp = [50, 100, 150, 200, 250] 
# extent of section and number of points
p_sect = [1000, 500] 
# x at fault bend, ramp angle, ramp height
p_ramp = [500, 30*np.pi/180, 100]
# total and incremental slip
p_slip = [300, 1]

# plot the fault bend fold
fault_bend_fold(yp, p_sect, p_ramp, p_slip)
```

Now let's add growth strata to the structure. In the cell below, we make the subsidence rate equal to the uplift rate:

```python
# plot the fault bend fold with growth strata
# and subsidence rate = uplift rate, G = 1.0
G = 1.0
fault_bend_fold(yp, p_sect, p_ramp, p_slip, G)
```

And below, we make the subsidence rate twice the uplift rate. Note how the growth strata geometries are segmented by kink axes. These are well explained in [Suppe et al. (1992)](https://link.springer.com/chapter/10.1007/978-94-011-3066-0_9).

```python
# plot the fault bend fold with growth strata
# and subsidence rate = uplift rate*2, G = 2.0
G = 2.0
fault_bend_fold(yp, p_sect, p_ramp, p_slip, G)
```

You can further experiment by modifying the angle of the fault ramp (careful, the maximum ramp angle for this model is 30), the height of the ramp, and the amount of slip.
