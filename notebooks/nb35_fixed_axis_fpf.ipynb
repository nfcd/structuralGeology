# Notebook 35: Fixed-axis fault propagation folds

Fault propagation folds consume slip at the tip of a propagating fault [(Suppe and Medwedeff, 1990)](https://www.e-periodica.ch/digbib/view?pid=egh-001%3A1990%3A83%3A%3A418&referrer=search#418). This notebook looks at the kink type, fixed axis fault propagation model. The figure below (from [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)) shows the kinematic model for a simple step, fixed axis fault propagation fold. Note that the model consists of three velocity domains, $v_1$ and $v_2$ are the velocities along the $\mathbf{x_1}$ and $\mathbf{x_2}$ axes, $s$ is the slip, and $R$ is a slip ratio (more details in [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)):

<img src="../figures/fixed_axis_fpf.png" alt="fixed_axis_fpf" width="500" style="display: block; margin: 0 auto"/><br><br>

The function [fixed_axis_fpf](../functions/fixed_axis_fpf.py) plots the evolution of a simple step, fixed axis fault propagation fold. Let's start by simulating a fold with a 20Â° dipping ramp. As you can see, this results in thickening of the forelimb.

```python
import numpy as np

# import our function
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "functions")))
from fixed_axis_fpf import fixed_axis_fpf

# define the input parameters
# beds datums
yp = [50, 100, 150, 200, 250] 
# extent of section and number of points
p_sect = [1000, 500] 
# x at fault bend, ramp angle, ramp height
# 20 deg dipping ramp -> forelimb thickening
p_ramp = [500, 20*np.pi/180] 
# total and incremental slip
p_slip = [100, 0.5]

# plot the fault propagation fold
fixed_axis_fpf(yp, p_sect, p_ramp, p_slip)
```

Now let's look at a fold above a 40Â° ramp. This results in forelimb thining.

```python
# 40 deg ramp -> forelimb thinning
p_ramp = [500, 40*np.pi/180] 
fixed_axis_fpf(yp, p_sect, p_ramp, p_slip)
```

The thickness of the forelimb is constant when the ramp angle is 29Â°. Let's add to this model growth strata.

```python
# 29 deg ramp -> forelimb thickness is constant
p_ramp = [500, 29*np.pi/180]
# subsidence rate = uplift rate * 2, G = 2.0
G = 2.0

fixed_axis_fpf(yp, p_sect, p_ramp, p_slip, G)
```

Finally, let's run a model with a fault ramp angle of 32Â°. This produces thinning of the forelimb.

```python
# 32 deg ramp 
p_ramp = [500, 32*np.pi/180]
# subsidence rate = uplift rate * 2, G = 2.0
G = 2.0

fixed_axis_fpf(yp, p_sect, p_ramp, p_slip, G)
```

Note how the growth strata geometries are segmented by kink axes. These are well explained in [Suppe et al. (1992)](https://link.springer.com/chapter/10.1007/978-94-011-3066-0_9).
