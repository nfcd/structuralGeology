# Notebook 37: Trishear fault propagation folds

In the kink type fault propagation folds covered before, the fold geometry is established at the start of the deformation, and the fold grows in a similar fashion by increasing in size but not changing in shape. This allows predicting fold geometry from fault geometry and vice versa. Howwever, kink models cannot explain features commonly observed in fault propagation folds, such as footwall synclines and thicknes and dip changes on forelimbs.

The trishear fault propagation fold model [(Allmendinger, 1998)](https://agupubs.onlinelibrary.wiley.com/doi/10.1029/98TC01907) explains that. In the trishear model, the displacement along the fault is accomodated by heterogeneous deformation in a triangular zone radiating from the fault tip (figure below from [Allmendinger et al., 2012](https://www.cambridge.org/core/books/structural-geology-algorithms/C5AB97C135925A5D49726B741738DCA9)):

<img src="../figures/trishear.png" alt="trishear" width="500" style="display: block; margin: 0 auto"/><br><br>

This produces richer fold geometries both in the hanging wall and footwall. The disadvantage though is that the model is incremental. Unlike kink type models, in the trishear model it is not possible to predict the fold geometry from the fault geometry.

The function [trishear](../functions/trishear.py) plots the evolution of a trishear fault propagation fold. Let's start by simulating a thrust related fold with a 30° dipping ramp.

```python
import numpy as np

# import our function
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "functions")))
from trishear import trishear

# define the input parameters
# beds datums
yp = [50, 80, 110, 140, 170] 
# extent of section and number of points
p_sect = [1000, 500] 
# trishear parameters
# x and y of fault tip, ramp angle, P/S,
# trishear angle, slip, and concentration factor
p_tri = [300, 50, 30*np.pi/180, 1.5, 60*np.pi/180, 100, 1.0]
# slip increment
sinc = 1.0

# plot the trishear fold
trishear(yp, p_sect, p_tri, sinc)
```

Let's add to this fold growth strata, we use a subsidence rate twice the uplift rate:

```python
# add growth strata
# subsidence rate = uplift rate * 2, G = 2.0
G = 2.0

# plot the trishear fold
trishear(yp, p_sect, p_tri, sinc, G)
```

As you can see, in the trishear model there is no instantaneous rotation of material across kink axes, but the growth strata are progressively rotated with each increment of folding.

We can use the same model for normal fault related deformation. In this case, both the fault slip and slip increment should be negative:

```python
# normal fault
yp = [150, 180, 210, 240, 270] 
# note that slip and slip increment must be negative
p_tri = [500, 50, 60*np.pi/180, 1.5, 60*np.pi/180, -100, 1.0]
sinc = -1.0

# plot the trishear fold
trishear(yp, p_sect, p_tri, sinc)
```

Finally, let's add growth strata. In this case, the subsidence rate is equal to the uplift rate:

```python
# add growth strata
# subsidence rate = uplift rate, G = 1.0
G = 1.0

# plot the trishear fold
trishear(yp, p_sect, p_tri, sinc, G)
```
