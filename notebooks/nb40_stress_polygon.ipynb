# Notebook 40: Stress polygon

This notebook is related to [notebook 31](nb31_faults_magnitude_stress.ipynb), and the possible stress states at a given depth and pore pressure, given that the stress in the crust is limited by the frictional strength of faults.

Jaeger and Cook (1979) showed that for a critically oriented fault at the frictional limit, the values of $\sigma_1$ and $\sigma_3$ are given by:

$$
R=\frac{\sigma_1}{\sigma_3}=\left[\left(\mu_s^2+1\right)^{1 / 2}+\mu_s\right]^2
$$

where $\mu_s$ is the coefficient of sliding friction. From Byerlees' law (Byerlee, 1978), we know $\mu_s$ has a narrow range, and a value of 0.6 is a reasonable estimate (for $\mu_s=0.6$, $R=3.1$).

We can use the equation above to estimate from a given vertical stress $\sigma_v$ and pore pressure $p$, the possible minimum and maximum horizontal stress values, $\sigma_{\mathrm{hmin}}$ and $\sigma_{\mathrm{hmax}}$, respectively (Zoback, 2010):

Normal faulting: 

$$\frac{\sigma_v-p}{\sigma_{\mathrm{hmin}}-p}=\left[\left(\mu_s^2+1\right)^{1 / 2}+\mu_s\right]^2$$

Strike-slip faulting:

$$\frac{\sigma_{\mathrm{hmax}}-p}{\sigma_{\mathrm{hmin}}-p}=\left[\left(\mu_s^2+1\right)^{1 / 2}+\mu_s\right]^2$$

Thrust faulting:

$$
\frac{\sigma_{\mathrm{hmax}}-p}{\sigma_v-p}=\left[\left(\mu_s^2+1\right)^{1 / 2}+\mu_s\right]^2
$$

In addition, the range of allowable values for the horizontal principal stresses in the Earth's crust for the normal, strike-slip and thrust fault regimes, can be illustrated in a stress polygon diagram (figure below from Zoback, 2010):

<img src="../figures/stress_polygon.png" alt="stress_polygon" width="500" style="display: block; margin: 0 auto"/><br><br>

As you can see, in a $\sigma_{\mathrm{hmin}}$ versus $\sigma_{\mathrm{hmax}}$ diagram, the plot displays the possible states of stress for normal faulting (NF), strike-slip faulting (SS), and reverse faulting (RF). These three are displayed as polygons, radiating from $\sigma_v$.

## Python function

The function [stress_polygon](../functions/stress_polygon.py) plot the Zoback's stress polygons for specified values of $\sigma_v$, $p$ and $\mu_s$ (the default $\mu_s = 0.6$).

## Application

Let's look at the stress polygons at a depth of about 1 km, $\sigma_v = 25$ MPa, $p=10$ MPa (hydrostatic), and the default $\mu_s=0.6$:

```python
import sys, os
sys.path.append(os.path.abspath(os.path.join("..", "functions")))
from stress_polygon import stress_polygon

sv = 25 # vertical stress = 25 MPa
pp = 10 # pore pressure = 10 MPa

stress_polygon(sv, pp)
```

Let's increase the pore pressure to 15 MPa and then to 20 MPa:

```python
# 15 MPa pore pressure
pp = 15

stress_polygon(sv, pp)
```

```python
# 20 MPa pore pressure
pp = 20

stress_polygon(sv, pp)
```

As you can see, elevated pore pressure reduces the difference between the principal stresses and the size of the polygons. When pore pressure is high, all principal stresses are close in magnitude to the vertical stress, and small changes in the stress field can cause a transition from one style of faulting to another (e.g. from NF to SS faulting) (Zoback, 2010).
